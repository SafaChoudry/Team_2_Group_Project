6. line chart visualizing the trend of total sales revenue over the four months

```python
y=df.dropna()
y["Sales_Revenue"]= ( y.Sales_Price * y.Quantity_Sold )
y
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name(locale='en_US')
yy=df.dropna()
yy["Sales_Revenue"]= ( df.Sales_Price * df.Quantity_Sold )
xx=yy.pivot_table(values='Sales_Revenue',index='Month',aggfunc='mean')
months=['January','February',"March","April"]
xx=xx.reindex(months)
xx.plot(kind='line', linewidth=3.0)
```

```python
jandf = df.loc[df['Month'] == "January"]
Jan = jandf['Quantity_Sold']*jandf['Sales_Price']
JAN = Jan.sum()
JAN
```

```python
febdf = df.loc[df['Month'] == "February"]
feb = febdf['Quantity_Sold']*febdf['Sales_Price']
FEB = feb.sum()
FEB
```

```python
mardf = df.loc[df['Month'] == "March"]
mar = mardf['Quantity_Sold']*mardf['Sales_Price']
MAR = mar.sum()
MAR
```

```python
aprdf = df.loc[df['Month'] == "April"]
apr = aprdf['Quantity_Sold']*aprdf['Sales_Price']
APR = apr.sum()
APR
```

```python
x = ["January", "February", "March", "April"]
y = [JAN,FEB,MAR,APR]

plt.plot(x, y)

plt.show()
```
